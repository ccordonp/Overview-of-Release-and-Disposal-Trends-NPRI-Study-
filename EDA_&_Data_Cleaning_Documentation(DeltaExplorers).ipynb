```python
continuous_columns = ['Release to Air - Fugitive',
                      'Release to Air - Other Non-Point ',
                      'Release to Air - Road dust  ',
                      'Release to Air - Spills ',
                      'Release to Air - Stack / Point  ',
                      'Release to Air - Storage / Handling ',
                      'Releases to Land - Leaks','Releases to Land - Other ',
                      'Releases to Land - Spills ',
                      'Releases to Water Bodies - Direct Discharges ',
                      'Releases to Water Bodies - Leaks',
                      'Releases to Water Bodies - Spills ',
                      'Off-site Disposal - Land Treatment',
                      'Off-site Disposal - Landfill',
                      'Off-site Disposal - Storage',
                      'Off-site Disposal - Tailings Management',
                      'Off-site Disposal - Underground Injection',
                      'Off-site Disposal - Waste Rock management',
                      'Off-site Transfers for Recycling - Energy Recovery',
                      'Off-site Transfers for Recycling - Other',
                      'Off-site Transfers for Recycling - Recovery of Acids and Bases',
                      'Off-site Transfers for Recycling - Recovery of Catalysts',
                      'Off-site Transfers for Recycling - Recovery of Inorganic Materials (not metals)',
                      'Off-site Transfers for Recycling - Recovery of Metals and Metal Compounds',
                      'Off-site Transfers for Recycling - Recovery of Organic Substances (not solvents)',
                      'Off-site Transfers for Recycling - Recovery of Pollution Abatement Residues',
                      'Off-site Transfers for Recycling - Recovery of Solvents',
                      'Off-site Transfers for Recycling - Refining or Re-use of Used Oil',
                      'Off-site Transfers for Treatment Prior to Final Disposal - Biological Treatment',
                      'Off-site Transfers for Treatment Prior to Final Disposal - Chemical Treatment',
                      'Off-site Transfers for Treatment Prior to Final Disposal - Incineration / Thermal',
                      'Off-site Transfers for Treatment Prior to Final Disposal - Municipal Sewage Treatment Plant',
                      'Off-site Transfers for Treatment Prior to Final Disposal - Physical Treatment',
                      'On-site Disposal - Land Treatment',
                      'On-site Disposal - Landfill',
                      'On-site Disposal - Tailings Management',
                      'On-site Disposal - Underground Injection',
                      'On-site Disposal - Waste Rock management']
```

```python
for column in continuous_columns: # looping through the continuous variables (substance columns)
    df_clean[column] = pd.to_numeric(df_clean[column], errors="coerce") # lets invalid data become missing value
    df_clean[column] = np.where(df_clean[column] < 0, 0, df_clean[column]) # makes negative values become 0


    pos = df_clean[df_clean[column] > 0][column]
    if len(pos) < 5: #skips columns with few positive values
        continue

    # using iqr to detect outliers
    q1 = pos.quantile(0.25)
    q3 = pos.quantile(0.75)
    iqr = q3 - q1

    if iqr <= 0:
        upper_bound = pos.quantile(0.99) # makes invalid values such as 0 or negative the 99th percentile to avoid invalid iqr detection
    else:
        upper_bound = q3 + 1.5 * iqr

    df_clean[column] = np.where(df_clean[column] > upper_bound, upper_bound, df_clean[column]) # we used the capping method, it reduces the extremity of a value, reference: https://www.analyticsvidhya.com/blog/2022/09/dealing-with-outliers-using-the-iqr-method/
```

```python
for column in continuous_columns:
    if df_clean[column].notna().sum() == 0: # skipping empty columns
        continue
    print(column,
          "min:", df_clean[column].min(),
          "max:", df_clean[column].max()) # checking max and min to see how they were affected by the outlier handling
```

```python
YEAR  = "Reporting_Year / Année" # year column
PROVINCE  = "PROVINCE" # province column
UNITS = "Units / Unités" # units column

df_time = df_clean.copy()
df_time[YEAR] = pd.to_datetime(df_time[YEAR], errors="coerce").dt.year
df_time = df_time[df_time[YEAR] >= 2012] # keeps the data from 2012-2022 to analyze the last 10 years only

# groups release by the same category
air_release = [
    "Release to Air - Fugitive",
    "Release to Air - Other Non-Point ",
    "Release to Air - Road dust  ",
    "Release to Air - Spills ",
    "Release to Air - Stack / Point  ",
    "Release to Air - Storage / Handling "
]

land_release = [
    "Releases to Land - Leaks",
    "Releases to Land - Other ",
    "Releases to Land - Spills "
]

water_release = [
    "Releases to Water Bodies - Direct Discharges ",
    "Releases to Water Bodies - Leaks",
    "Releases to Water Bodies - Spills "
]
# gets the sum of each group and the total release for visualization comparison
df_time["AIR"] = df_time[air_release].sum(axis=1)
df_time["LAND"] = df_time[land_release].sum(axis=1)
df_time["WATER"] = df_time[water_release].sum(axis=1)
df_time["TOTAL"] = df_time["AIR"] + df_time["LAND"] + df_time["WATER"]

# groups data by year, province, and the total releases
by_year = df_time.groupby([YEAR, PROVINCE])["TOTAL"].sum().reset_index()
```

```python
import matplotlib.pyplot as plt # importing matplotlib for visualizations
import seaborn as sns

units = ["grams", "kg", "tonnes", "g TEQ"] # unit values in the data

for unit in units: # for every unit in the list
    group = df_time[df_time[UNITS] == unit] # creates a group of values with the same unit

    by_year = group.groupby([YEAR, PROVINCE])["TOTAL"].sum().reset_index() # total release by year and province
    top5 = by_year[[PROVINCE, "TOTAL"]].groupby(PROVINCE).sum().sort_values("TOTAL", ascending=False)[:5].index # sorting it by descending order to fet the top 5 province through the yeara
    plot_data = by_year[by_year[PROVINCE].isin(top5)]
    plt.figure(figsize=(9,4.5))
    sns.lineplot(data=plot_data, x=YEAR, y="TOTAL", hue=PROVINCE, marker="o") # plot total releases by province over time with o (circles) as markers

    plt.title(f"Releases by Year — Top 5 Provinces ({unit})")
    plt.xlabel("Year")
    plt.ylabel(f"Total release ({unit})")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
YEAR  = "Reporting_Year / Année" # year column
COMPANIES  = "Company_Name / Dénomination_sociale_de_l'entreprise" # companies column
UNITS = "Units / Unités" # units column

# only keep 2012 until 2022
df_time = df_time[df_time[YEAR] >= 2012]

units = ["grams", "kg", "tonnes", "g TEQ"] # creates a list for different value in units

for unit in units: # loop through each unit
    group = df_time[df_time[UNITS] == unit] # get data from a specific unit

    # groups data by company, total releases, and sorts them in descending order to get the top 10
    top10 = group.groupby(COMPANIES)["TOTAL"].sum().sort_values(ascending=False)[:10].index
    plot_data = group[group[COMPANIES].isin(top10)] # keeps the top 10 companies only

    plt.figure(figsize=(9, 5))
    plt.barh(
        plot_data.groupby(COMPANIES)["TOTAL"].sum().sort_values(ascending=True).index, # index stands for the company names
        plot_data.groupby(COMPANIES)["TOTAL"].sum().sort_values(ascending=True).values # returns the values (releases) of companies
    )
    plt.title(f"Top 10 Companies by Total Releases ({unit}, 2012–2022)")
    plt.xlabel(f"Total releases ({unit})") # total releases of units as the x label
    plt.tight_layout()
    plt.show()
```

```python
units = ["grams", "kg", "tonnes", "g TEQ"]

for unit in units: # loop through every unit
    data = df_time[df_time[UNITS] == unit] # rows that match a specific unit

    # total air,land,water releases by year and province
    grouped = data.groupby([PROVINCE, YEAR])[["AIR", "LAND", "WATER"]].sum().reset_index()

    start_year = grouped[YEAR].min() # first year of the dataset
    end_year = grouped[YEAR].max() # last year of the dataset

    # data for first and last year by province
    start = grouped[grouped[YEAR] == start_year].set_index(PROVINCE)
    end = grouped[grouped[YEAR] == end_year].set_index(PROVINCE)

    merged = start.join(end, lsuffix="_start", rsuffix="_end").fillna(0) # join start and end together, reference:https://www.datacamp.com/tutorial/joining-dataframes-pandas

    for col in ["AIR", "LAND", "WATER"]:
        merged[f"{col}_drop"] = merged[f"{col}_start"] - merged[f"{col}_end"] # calculate the decrease of the groups

        # top 5 largest decrease
        top5 = merged.sort_values(f"{col}_drop", ascending=False).head(5).reset_index()

        plt.figure(figsize=(6, 3))
        sns.barplot(data=top5, x=PROVINCE, y=f"{col}_drop", color="lightcoral")
        plt.title(f"Top 5 {col} decreases ({unit}, {start_year}-{end_year})")
        plt.xlabel("Province")
        plt.ylabel(f"Decrease ({unit})")
        plt.tight_layout()
        plt.show()
```
